```python
import pandas as pd
```

```python
######################## 데이터 병합 - concat 사용
```

```python
df1=pd.DataFrame({'A': ['A0', 'A1', 'A2', 'A3']
                 , 'B': ['B0', 'B1', 'B2', 'B3']
                 , 'C': ['C0', 'C1', 'C2', 'C3']
                 , 'D': ['D0', 'D1', 'D2', 'D3']}
                 , index=[0,1,2,3])

df2=pd.DataFrame({'A': ['A0', 'A1', 'A2', 'A3']
                 , 'B': ['B0', 'B1', 'B2', 'B3']
                 , 'C': ['C0', 'C1', 'C2', 'C3']
                 , 'D': ['D0', 'D1', 'D2', 'D3']}
                 , index=[0,1,2,3])

df3=pd.DataFrame({'A': ['A0', 'A1', 'A2', 'A3']
                 , 'B': ['B0', 'B1', 'B2', 'B3']
                 , 'C': ['C0', 'C1', 'C2', 'C3']
                 , 'D': ['D0', 'D1', 'D2', 'D3']}
                 , index=[0,1,2,3])
```

```python
df1
```

```python
df2
```

```python
df3
```

```python
# 열 방향으로 병합
result = pd.concat([df1, df2, df3])
result
```

```python
# 열 방향으로 병합할때 key 값을 부여
result = pd.concat([df1, df2, df3], keys=['x','y','z'])
result
```

```python
result.index
```

```python
result.index.get_level_values(0)
```

```python
result.index.get_level_values(1)
```

```python
df4 = pd.DataFrame({
                    'B': ['B2','B3','B6','B7'],
                    'D': ['D2','D3','D6','D7'],
                    'F': ['F2','F3','F6','F7']
                   }, index=[2,3,6,7])

df1
```

```python
df4
```

```python
result = pd.concat([df1,df4])
result
```

```python
# axis 를 주면 index 를 기준으로 병합, 안주면 그냥 병합하여 index 가 섞인다.
result = pd.concat([df1,df4], axis=1)
result
```

```python
# index로 병합되지 않는 것은 제거하는 옵션 => join='inner'
result = pd.concat([df1,df4], axis=1, join='inner')
result
```

```python
# 한쪽 기준으로 인덱스 맞춤
result = pd.concat([df1, df4], axis=1, join_axes=[df1.index])
result
```

```python
# 기존의 index를 무시하고 새로 index 부여
result = pd.concat([df1, df4], ignore_index=True)
result
```

```python
######################## 데이터 병합 - merge
```

```python
left = pd.DataFrame({'key': ['K0', 'K4', 'K2', 'K3'],
                     'A': ['A0', 'A1', 'A2', 'A3'],
                     'B': ['B0', 'B1', 'B2', 'B3']})

right = pd.DataFrame({'key': ['K0', 'K1', 'K2', 'K3'],
                     'C': ['C0', 'C1', 'C2', 'C3'],
                     'D': ['D0', 'D1', 'D2', 'D3']})
```

```python
left
```

```python
right
```

```python
# 공통된 key 에 대해서만 합치게 됨
pd.merge(left, right, on='key')
```

```python
pd.merge(left, right, how='left', on='key')
```

```python
pd.merge(left, right, how='right', on='key')
```

```python
pd.merge(left, right, how='outer', on='key')
```

```python
pd.merge(left, right, how='inner', on='key')
```
