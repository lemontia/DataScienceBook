```python
import pandas as pd
```

```python
df1=pd.DataFrame({'A': ['A0', 'A1', 'A2', 'A3']
                 , 'B': ['B0', 'B1', 'B2', 'B3']
                 , 'C': ['C0', 'C1', 'C2', 'C3']
                 , 'D': ['D0', 'D1', 'D2', 'D3']}
                 , index=[0,1,2,3])

df2=pd.DataFrame({'A': ['A0', 'A1', 'A2', 'A3']
                 , 'B': ['B0', 'B1', 'B2', 'B3']
                 , 'C': ['C0', 'C1', 'C2', 'C3']
                 , 'D': ['D0', 'D1', 'D2', 'D3']}
                 , index=[0,1,2,3])

df3=pd.DataFrame({'A': ['A0', 'A1', 'A2', 'A3']
                 , 'B': ['B0', 'B1', 'B2', 'B3']
                 , 'C': ['C0', 'C1', 'C2', 'C3']
                 , 'D': ['D0', 'D1', 'D2', 'D3']}
                 , index=[0,1,2,3])
```

```python
df1
```

```python
df2
```

```python
df3
```

```python
# 열 방향으로 병합
result = pd.concat([df1, df2, df3])
result
```

```python
# 열 방향으로 병합할때 key 값을 부여
result = pd.concat([df1, df2, df3], keys=['x','y','z'])
result
```
